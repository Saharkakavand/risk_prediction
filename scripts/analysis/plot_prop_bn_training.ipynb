```julia
using AutoRisk
using Discretizers
using DataFrames
using JLD
using PGFPlots
```

```julia
filepath = "../../data/results/data/stats.jld"
d = JLD.load(filepath)
stats = d["stats"]
```

```julia
g = GroupPlot(2, 2, groupStyle = "horizontal sep = 1cm, vertical sep = 1.5cm");
itr = length(stats["aggressiveness"])

a = Axis(
Plots.Linear(collect(1:itr), stats["aggressiveness"], mark="none"), 
    width="5cm",
    height="4cm",
    title="Aggressiveness",
)
push!(g, a)

a = Axis(
Plots.Linear(collect(1:itr), stats["isattentive"], mark="none"), 
    width="5cm",
    height="4cm",
    title="Attentiveness",
)
push!(g, a)

a = Axis(
Plots.Linear(collect(1:itr), stats["relvelocity"], mark="none"), 
    width="5cm",
    height="4cm",
    title="Relative Velocity",
    xlabel="Iterations"
)
push!(g, a)

a = Axis(
Plots.Linear(collect(1:itr), stats["utilities"], mark="none"), 
    width="5cm",
    height="4cm",
    title="Collision Probability",
    xlabel="Iterations"
)
push!(g, a)
```

```julia
PGFPlots.save("../../data/results/vis/cem.tex", g)
```
